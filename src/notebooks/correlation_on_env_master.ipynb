```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

env = pd.read_parquet(
    "../../data/curated/environmental_master.parquet"
)
```

```python
features = [
    "aqi",
    "pm25",
    "pm10",
    "humidity",
    "temperature",
    "temperature_2m",
    "relative_humidity_2m",
    "dew_point_2m",
    "precipitation",
    "surface_pressure",
    "cloud_cover",
    "wind_speed_10m",
    "wind_direction_10m",
    "wind_gusts_10m"
]

corr_df = env[features].copy()
```

```python
corr_matrix = corr_df.corr(
    method="pearson"
)

corr_matrix.round(3)
```

```python
plt.figure(figsize=(14,12))

sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    center=0,
    square=True,
    linewidths=0.5
)

plt.title(
    "Feature-to-Feature Correlation Matrix",
    fontsize=16
)

plt.tight_layout()

plt.show()
```

```python
corr_matrix.to_csv(
    "../../src/reports/full_feature_correlation_matrix.csv"
)

print(
    "Saved: full_feature_correlation_matrix.csv"
)
```
